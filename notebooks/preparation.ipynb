# Parameters
- diff_c: diffusion constant in μm/s²
- r: focus feedback range in μm
- dr: focus feedback error in μm
- dt: time lapse time interval in s
- p: probability each frame in focus
- t/total_time: experiment time

```python
# find max time interval in seconds where each frame is in focus with certain p
focus_probability.get_dt(t=1000, diff_c=2e-4, r=0.3, dr=0.05, p=0.95)
```

```python
# find max diffusion constant in um2/s where each frame is in focus with certain p
focus_probability.get_diff_c(t=1000, dt=0.1, r=0.3, dr=0.05, p=0.95)
```
